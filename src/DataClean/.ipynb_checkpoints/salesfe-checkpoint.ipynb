```python
import pandas as pd
```

```python
files = ['./Data/Rolling/ManhattanSalesFinal.csv',
    './Data/Rolling/QueensSalesFinal.csv',
    './Data/Rolling/StatenIslandSalesFinal.csv',
    './Data/Rolling/BrooklynSalesFinal.csv',
    './Data/Rolling/BronxSalesFinal.csv']
```

```python
ffiles= ['./Data/Rolling/ManhattanSalesFinalFE.csv',
    './Data/Rolling/QueensSalesFinalFE.csv',
    './Data/Rolling/StatenIslandSalesFinalFE.csv',
    './Data/Rolling/BrooklynSalesFinalFE.csv',
    './Data/Rolling/BronxSalesFinalFE.csv']
```

```python
fileNum = 4
```

```python
df = pd.read_csv(files[fileNum])
```

```python
df.head(30)
```

```python
df['h3geo'] = 0
```

```python
df['YEAR BUILT'].describe()
```

```python
for i in range(1800, 2030, 10):
    df['{:4.0f}s'.format(i)] = 0
    
for index, row in df.iterrows():
    col = row['YEAR BUILT'] - (row['YEAR BUILT'] % 10)
    df.iloc[index, df.columns.get_loc('{:4.0f}s'.format(col))] = 1
```

```python
df = pd.get_dummies(df, prefix=['BUILDING CLASS CATEGORY','BUILDING CLASS','TAX CLASS'], columns=['BUILDING CLASS CATEGORY','BUILDING CLASS AT PRESENT', 'TAX CLASS AT TIME OF SALE'])
```

```python
df.to_csv(ffiles[fileNum], index=False)
```

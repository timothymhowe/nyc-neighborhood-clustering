```python
from shapely.geometry import Polygon

def convert_sublist_to_tuple(the_list):
    if type(the_list[0]) != list:
        a , b = the_list
        return a,b
        # return str(a) , str(b)
    ret_list = the_list
    for i in range(0,len(ret_list)):
        ret_list[i] = convert_sublist_to_tuple(the_list[i])
    
    return ret_list

def convert_sublist_to_polygon(the_list):
    if type(the_list[0]) != list:
        return shapely.geometry.Polygon(the_list)
    ret_list = the_list
    for i in range(0,len(ret_list)):
        ret_list[i] = convert_sublist_to_polygon(the_list[i])
    
    return ret_list

```

```python
import shapely
from shapely.geometry import MultiPolygon, Point

def get_hex_boundary(boro_hex_set_u):
    boro_helper = pd.DataFrame([geojson.Polygon(h3.h3_set_to_multi_polygon(boro_hex_set_u))])
    boro_helper = convert_sublist_to_tuple(list(boro_helper['coordinates']))
    boro_helper = convert_sublist_to_polygon(boro_helper)[0]
    boro_helper = list(map(shapely.geometry.MultiPolygon,boro_helper))
    # print(len(boro_helper[0]))
    boro_helper = [x for x in boro_helper if x.area >= .001]
    
    out_gdf = gpd.GeoDataFrame()
    out_gdf['geometry'] = boro_helper
    out_gdf['geometry_long_lat'] = gpd.GeoSeries(out_gdf['geometry']).map(lambda polygon: shapely.ops.transform(lambda x, y: (y, x), polygon))
    out_gdf_lon_lat = out_gdf.drop('geometry',axis=1)
    out_gdf_lon_lat['geometry'] = out_gdf_lon_lat['geometry_long_lat']
    
    out_gdf_finished = out_gdf_lon_lat.set_crs("EPSG:4326")
    
    return out_gdf_finished


queens_hex_boundary = get_hex_boundary(queens_hex_set_u)
brooklyn_hex_boundary = get_hex_boundary(brooklyn_hex_set_u)
staten_hex_boundary = get_hex_boundary(staten_hex_set_u)
manhattan_hex_boundary = get_hex_boundary(manhattan_hex_set_u)
bronx_hex_boundary = get_hex_boundary(bronx_hex_set_u)



# brooklyn_hex = pd.DataFrame(brooklyn_hex)
# brooklyn_hex_int = brooklyn_hex[0].apply(lambda x: h3.string_to_h3(x))
# # queens_test = pd.DataFrame([geojson.Polygon(h3.h3_set_to_multi_polygon(queens_hex_set_u))])
# queens_test = pd.DataFrame([geojson.Polygon(h3.h3_set_to_multi_polygon(queens_hex_set_u))])

# mayb = convert_sublist_to_tuple(list(queens_test['coordinates']))
# bayb = convert_sublist_to_polygon(mayb)[0]
# # queens_test = gpd.GeoSeries(pd.DataFrame(queens_test)['coordinates'].map(lambda polygon: shapely.ops.transform(lambda x, y: (x, y), polygon)))
# # queens_test

# gleyb = list(map(shapely.geometry.MultiPolygon,bayb))
# queens_test['coordinates'] = mayb
# queens_test['coordinates'].apply(lambda x : Polygon(evalx))

# df.loc[df['coordinates'].str.len()>=3, 'coordinates'].apply(Polygon)
# lines = []
# for pol in gleyb:
#     boundary = pol.boundary
#     if boundary.type == 'MultiLineString':
#         for line in boundary:
#             lines.append(line)
#     else:
#         lines.append(boundary)
# f_yeah = gpd.GeoDataFrame()
# f_yeah['geometry'] = gleyb
# f_yeah.plot()
# for line in lines
#     print(line.wkt
# f_yeah
```

```python
from h3 import h3
import folium

def visualize_hexagons(hexagons, color="red", folium_map=None):
    """
    hexagons is a list of hexcluster. Each hexcluster is a list of hexagons. 
    eg. [[hex1, hex2], [hex3, hex4]]
    """
    polylines = []
    lat = []
    lng = []
    for hex in hexagons:
        polygons = h3.h3_set_to_multi_polygon([hex], geo_json=False)
        # flatten polygons into loops.
        outlines = [loop for polygon in polygons for loop in polygon]
        polyline = [outline + [outline[0]] for outline in outlines][0]
        lat.extend(map(lambda v:v[0],polyline))
        lng.extend(map(lambda v:v[1],polyline))
        polylines.append(polyline)
    
    if folium_map is None:
        # m = folium.Map(location=[sum(lat)/len(lat), sum(lng)/len(lng)], zoom_start=13, tiles='cartodbpositron',prefer_canvas=True)
        m = folium.Map(location=[sum(lat)/len(lat), sum(lng)/len(lng)], zoom_start=11)
    else:
        m = folium_map
    for polyline in polylines:
        my_PolyLine=folium.PolyLine(locations=polyline,weight=2,color=color)
        m.add_child(my_PolyLine)
    return m
    

def visualize_polygon(polyline, color="red",folium_map=None):
    polyline.append(polyline[0])
    lat = [p[0] for p in polyline]
    lng = [p[1] for p in polyline]
    if folium_map is None:
        m = folium.Map(location=[sum(lat)/len(lat), sum(lng)/len(lng)], tiles='cartodbpositron')
    else:
        m = folium_map
    my_PolyLine=folium.PolyLine(locations=polyline,weight=2,color=color)
    m.add_child(my_PolyLine)
    return m
```

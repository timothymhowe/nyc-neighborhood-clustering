```python
import numpy as np
import pandas as pd
import geopandas as gpd
import h3pandas

neigh_gpd = gpd.read_file(
    'https://services5.arcgis.com/GfwWNkhOj9bNBqoJ/arcgis/rest/services/Neighborhood_Names/FeatureServer/0/query?where=1=1&outFields=*&outSR=4326&f=pgeojson')
```

```python
neigh_gpd = neigh_gpd.h3.geo_to_h3(10)
neigh_gpd = neigh_gpd[neigh_gpd['Stacked'] < 3]
```

```python
neigh_gpd.reset_index()[['h3_10', 'Name']].to_csv('../../data/clean/neighborhood_centroids.csv', index=False)
```
